```python
data.info()
```

```python
data.dropna(inplace=True)
data.isnull().sum()
```

```python
prices = list(data['Price'])
non_numeric = []
for i,val in enumerate(prices):
    try:
        int(val)
    except:
        non_numeric.append(val)
        print(val, end=" ")
```

```python
# Dropping POA from dataset
for non_num in non_numeric:
    data = data[data['Price'] != non_num]
    
# Converting price column from object to int type
data['Price_int'] = data['Price'].apply(lambda x: int(x))
data.info()
```

<h1> #2 Make and Exploratory Data Analysis (EDA)

```python
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
brands = data['Brand']
unique_brands = np.unique(brands) # создал список уникальных брендов

brand_counts = dict(data['Brand'].value_counts()) # бренды и их количество в датасет
print(brand_counts)
```

```python
x = list(brand_counts.keys())[:10]
height = list(brand_counts.values())[:10]
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
sns.barplot(x=x, y=height, palette="viridis")
plt.title("Top 10 Cars with Count", fontsize=18)
plt.xlabel("Brand Names", fontsize=14)
plt.ylabel("Count of Cars", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.show()

# тут я создал график в котором показывают топ 10 брендов машин (брал с предыдущего ) и расставил по х названия, по у количество
```

```python
fuel_consumption = data['FuelConsumption']

# Создание списка пробега и добавление среднего значения 'mean', где l == 0
mileage = []
for val in fuel_consumption:
    l = val.split("/")[0]
    l = float(l.split(' ')[0])
    if l == 0: # добавляю минимум значение if l == 0
        l = 'mean'
        mileage.append(l)
    else:
        mileage.append(round(100/l))

# Получение индексов для всех средних значений 'mean' 
mean_ind = []
for i,val in enumerate(mileage):
    if val == 'mean':
        mean_ind.append(i)

# Удаление среднего значения, обусловленного малым значением L
for val in mileage:
    try:
        mileage.remove('mean')
    except:
        pass
    
# Добавление совокупных значений набора данных в индекс средних значений
for ind in mean_ind:
    mileage.insert(ind, sum(mileage)//len(mileage))
```

```python
data['Mileage'] = mileage
data.sample(3)
```
